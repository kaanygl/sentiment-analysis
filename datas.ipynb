```python
columns = list(df)

conn = sqlite3.connect('election.db') 

cursor = conn.cursor() 

ids= []

for k in range(len(df)):
    ids.append(k)
    elements = []
    for i in columns:
        element = df[i][k]
        
        if isinstance(element, np.integer):
            #sqlite3.register_adapter(np.int64, lambda element: int(element))
            element = element.tolist()
        elements.append(element)
    
    cursor.execute('''INSERT INTO datas(id, Analysis_Description, Total_Analysed_Data, Topic_Analysis, Side_A, Side_B, Side_C, Sentiment_Score_of_A, Sentiment_Score_of_B,
                Percentage_of_A, Percentage_of_B, Percentage_of_C, Total_Percentage) 
                VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?)''', (ids[k], elements[0], elements[1],elements[2], elements[3],elements[4], elements[5],elements[6], elements[7],elements[8], elements[9],elements[10], elements[11]))


    


conn.commit() 
  
# Closing the connection 
conn.close()
    
   
```
